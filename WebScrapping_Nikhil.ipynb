```python
from selenium import webdriver
import time

# specifies the path to the chromedriver.exe
browser = webdriver.Chrome('C:/Users/nikhi/Downloads/Web Scrapping/drivers/chromedriver.exe')
# driver.get method() will navigate to a page given by the URL address
browser.get('https://www.pizzahut.com.au/stores')
time.sleep(5)
```

```python
import pandas as pd

storeContent = browser.find_elements_by_class_name("store-content")

df_stores = pd.DataFrame(columns=['StoreName', 'StoreAddress', 'URL'])

for item in storeContent:
    name = item.find_element_by_class_name("store-name").text
    address = item.find_element_by_class_name("store-address").text
    url = item.find_element_by_xpath("a").get_attribute("href")
    
    
    new_row = {'StoreName':name, 'StoreAddress': address, 'URL':url}
    #append row to the dataframe
    df_stores = df_stores.append(new_row, ignore_index=True)
    
df_stores.head(10)
```

```python
df_stores.to_csv("pizzahut_stores.csv")
```

```python
from selenium.webdriver.common.keys import Keys

df = pd.read_csv("pizzahut_stores.csv")
urls = df["URL"]
location = []
address = []

# specifies the path to the chromedriver.exe
browser = webdriver.Chrome('C:/Users/nikhi/Downloads/Web Scrapping/drivers/chromedriver.exe')
# driver.get method() will navigate to a page given by the URL address
browser.get('https://www.pizzahut.com.au/stores')
time.sleep(2)


main_window = browser.current_window_handle

for url in urls:
    browser.find_element_by_tag_name('body').send_keys(Keys.CONTROL + Keys.TAB)
    
    # driver.get method() will navigate to a page given by the URL address
    browser.get(url)
    
    # Put focus on current window which will, in fact, put focus on the current visible tab
    browser.switch_to_window(main_window)
    
    # Close current tab
    browser.find_element_by_tag_name('body').send_keys(Keys.CONTROL + 'w')
    
    time.sleep(2)
    
    reqDiv = browser.find_elements_by_tag_name("span")
    if not reqDiv:
        continue
    else:
        temp = reqDiv[0]
        temp = temp.text
        location.append(temp)
        reqDiv = reqDiv[-7:-3]
    
        element = ""
        for item in reqDiv:
            element = element + item.text +", "
        
        element = element[:-3]
        address.append(element)
```

```python
import pandas as pd

dictLocation = {'Location': location, 'Address': address}
df = pd.DataFrame(dictLocation)
df.head(20)
```

```python
df.to_csv("store_details.csv")
```
